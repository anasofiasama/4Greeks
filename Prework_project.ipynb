#### Make a plot of the coordinates (lat and long) of the south belt of Madrid by color of each population (★★★★) 
This is a extremely hard exercise and we highly recommend to do it at the end of everything

```python
# Run in console the following command before run the rest of code:
# jupyter nbextension enable --py --sys-prefix ipyleaflet
#!pip install ipyleaflet
from ipyleaflet import Map, basemaps

# Map centred on (60 degrees latitude et -2.2 degrees longitude)
# Latitude, longitude
map = Map(center = (60, -2.2), zoom = 2, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.Stamen.Terrain)
map
```

Para crear los marcadores las coordenadas deben ser numericas, por lo tanto primero debo convertir las columnas de longitude y latitude a numericas.

```python
sdf['longitude']=sdf['longitude'].str.replace(',','.')
sdf['latitude']=sdf['latitude'].str.replace(',','.')

sdf[['latitude','longitude']]=sdf[['latitude','longitude']].applymap(float)
sdf.info()
```

```python
# https://coderzcolumn.com/tutorials/data-science/ipyleaflet-interactive-maps-in-python-based-on-leafletjs
    
## HERE: plot the coordinates of the estates
## PUT HERE YOUR CODE:
from ipyleaflet import Marker,MarkerCluster, AwesomeIcon

group={}
color=['green','red','blue','yellow']
for name, city, lon, lat in sdf[["index",'level5',"longitude", "latitude"]].values:
    marker = Marker(location=(lat, lon), draggable=False, title=str(name), alt=str(name))
    if city not in group:
        group[city]=list()
        group[city].append(marker)
    else:
        group[city].append(marker)
        
for key,value in group.items():
    map.add_layer(MarkerCluster(markers=value))
```

#### What can you say about the price per square meter between the towns of Getafe and Alcorcón?

```python
sdf['priceperm']=sdf['price']/sdf['surface']
sdf[(sdf['level5']=='Getafe')|(sdf['level5']=='Alcorcón')].groupby('level5')['priceperm'].hist(alpha=0.5)
plt.title('Precio por metro cuadrado por ciudad')
plt.xlabel('Precio por metro cuadrado')
plt.legend(['Alcorcon','Getafe'])
plt.show()
```

```python
sdf[(sdf['level5']=='Getafe')|(sdf['level5']=='Alcorcón')].groupby('level5')['priceperm'].mean()
```

Del grafico se observa que en Getafe existen algunos valores muy bajos (posibles outliers). Seguramente estos incidan en que en promedio el precio medio por metro cuadrado sea menor en Getafe que en Alcorcon.
